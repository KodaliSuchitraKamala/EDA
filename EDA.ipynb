# 📊 Exploratory Data Analysis (EDA)

**EDA** is the first step in data analysis, helping you uncover insights and identify key patterns in your dataset. It allows you to explore what questions to ask before diving into detailed analysis.

This process can be done independently or alongside data cleaning.

### During EDA, you will:
- Ask meaningful questions about the data
- Identify the underlying structure of the dataset
- Detect trends, patterns, and anomalies
- Test hypotheses and validate assumptions
- Explore potential problems that can be solved using the data

```python
# ✅ Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# 📥 Load sample dataset: Titanic from seaborn
df = sns.load_dataset('titanic')
df.head()
```

```python
# 🔍 Basic information and data types
df.info()
```

```python
# 📊 Descriptive statistics
df.describe(include='all')
```

```python
# 🧹 Check for missing values
df.isnull().sum()
```

```python
# 📈 Visualizing distributions
plt.figure(figsize=(10,5))
sns.histplot(df['age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# 🎯 Survival rate by gender
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# 🔥 Correlation heatmap (numerical features only)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```
